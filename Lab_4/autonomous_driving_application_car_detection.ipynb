### Loading and Running YOLOv8

In this section, we load a lightweight **pretrained YOLOv8 model** (`yolov8n.pt`) provided by Ultralytics.  
This model is trained on the COCO dataset and is suitable for fast inference on common objects, including cars.  

The helper function `run_yolov8` is defined to:  
- Run YOLO inference on a given image.  
- Apply a **confidence threshold** (default 0.25) to filter low-confidence detections.  
- Apply an **IoU threshold** (default 0.45) to control Non-Maximum Suppression (removing overlapping boxes).  
- Return three arrays:  
  - Bounding boxes in `(y1, x1, y2, x2)` format.  
  - Confidence scores for each detection.  
  - Predicted class IDs.  

This function standardizes outputs, making them easy to use for further analysis and visualization later in the lab.


````python
from ultralytics import YOLO # Import YOLO from ultralytics

# Load a small pretrained pytorch model
model = YOLO("yolov8n.pt") # Load the yolov8n.pt model
print("Loaded YOLOv8 model") # Print a confirmation message

def run_yolov8(image_path, score_threshold=0.25, iou_threshold=0.45):
    # Run YOLO inference on the image
    results = model.predict(image_path, conf=score_threshold, iou=iou_threshold, verbose=False)
    # Check if results are empty or no boxes are detected
    if not results or results[0].boxes is None:
        return np.zeros((0,4), np.float32), np.zeros((0,), np.float32), np.zeros((0,), np.int32)
    r = results[0] # Get the first result
    xyxy = r.boxes.xyxy.cpu().numpy().astype(np.float32)  # (x1,y1,x2,y2) - Get bounding box coordinates
    conf = r.boxes.conf.cpu().numpy().astype(np.float32) # Get confidence scores
    cls  = r.boxes.cls.cpu().numpy().astype(np.int32) # Get class IDs
    yx = xyxy[:, [1,0,3,2]]  # → (y1,x1,y2,x2) - Convert coordinates to (y1, x1, y2, x2) format
    return yx, conf, cls # Return bounding boxes, confidence scores, and class IDs
````

### Drawing Detection Boxes on Images

The function `draw_boxes` is used to visualize YOLO detections directly on images.  
It takes as input the image path, bounding boxes, confidence scores, and class labels, and then overlays them on the image.  

Key features of this function:  
- **Bounding box drawing**: Green rectangles are drawn around detected objects.  
- **Labels**: Each box is annotated with the class name (or ID) and its confidence score.  
- **Filtering options**:  
  - `min_score` filters out low-confidence predictions.  
  - `wanted_idx` allows drawing only a specific class (e.g., cars only).  
- **Font handling**: Attempts to use a TrueType font if available, otherwise defaults to system font.  
- **Saving results**: If `save_back=True`, the function saves the annotated image back to disk with a suffix (e.g., `_detected`).  

The function returns:  
1. The annotated `PIL.Image` object.  
2. The number of drawn detections.  
3. The file path of the saved image (if saving is enabled).  

This step provides a convenient way to **inspect detection quality visually** and generate results for reports.


````python
from PIL import ImageDraw, ImageFont # Import ImageDraw and ImageFont from PIL

from PIL import ImageDraw, ImageFont # Import ImageDraw and ImageFont from PIL again (redundant)

def draw_boxes(image_path, boxes, scores, classes,
               class_names=None, wanted_idx=None,
               min_score=0.25, font_path="font/arial.ttf",
               save_back=False, suffix="_detected"):
    """
    Draw boxes on image, optionally overwrite/save back into images folder.
    """
    img = Image.open(image_path).convert("RGB") # Open the image and convert to RGB
    draw = ImageDraw.Draw(img) # Create a drawing object
    try:
        font = ImageFont.truetype(font_path, 14) # Try to load a TrueType font
    except:
        font = None # If font loading fails, set font to None

    count = 0 # Initialize detection count
    for b, s, c in zip(boxes, scores, classes): # Iterate through boxes, scores, and classes
        if s < min_score: # Skip if confidence score is below the minimum
            continue
        if wanted_idx is not None and c != wanted_idx: # Skip if class ID does not match the wanted index
            continue
        y1, x1, y2, x2 = [int(v) for v in b] # Get integer coordinates of the bounding box
        draw.rectangle([x1, y1, x2, y2], outline=(0,255,0), width=3) # Draw the rectangle
        label = f"{class_names[c] if class_names else c}: {s:.2f}" # Create the label text
        draw.text((x1+3, max(0, y1-12)), label, fill=(255,255,0), font=font) # Draw the label text
        count += 1 # Increment the detection count

    # save back into same folder
    if save_back: # If save_back is True
        base, ext = os.path.splitext(image_path) # Split the file path into base and extension
        new_path = f"{base}{suffix}{ext}"   # e.g. 0114_detected.jpg - Create the new file path
        img.save(new_path) # Save the annotated image
        print(f"Saved annotated image: {new_path}") # Print a confirmation message
        return img, count, new_path # Return the image, count, and new path

    return img, count # Return the image and count
````

### Post-Processing with TensorFlow

The helper function `postprocess_with_tf2` refines the raw YOLOv8 detections using TensorFlow utilities for bounding box handling.  
It allows consistent filtering, class selection, and non-maximum suppression (NMS) in a **TensorFlow-friendly pipeline**.

Steps inside the function:
1. **Run YOLOv8**: Calls `run_yolov8` to get raw detections (boxes, confidence, class IDs).  
2. **Tensor Conversion**: Converts confidence scores and bounding boxes to TensorFlow tensors.  
3. **One-Hot Encoding of Classes**: Builds a one-hot representation of predicted classes for TensorFlow operations.  
4. **Bounding Box Post-Processing**: Uses `yolo_eval` to apply:  
   - Confidence threshold filtering.  
   - IoU threshold filtering (Non-Maximum Suppression).  
   - Per-class NMS to handle overlapping boxes of the same object.  
5. **Class Filtering**: If a `wanted_class` is specified (e.g., `"car"`), detections are restricted to that class only.  
6. **Returns**:  
   - Final confidence scores.  
   - Refined bounding boxes in `(y1, x1, y2, x2)` format.  
   - Predicted class IDs.

This post-processing step is useful for comparing **YOLO’s built-in handling** with a **TensorFlow-based approach**, and for seamlessly integrating detections into downstream TensorFlow models or pipelines.


````python
def postprocess_with_tf2(image_path, score_threshold=0.25, iou_threshold=0.5, max_boxes=50, wanted_class="car"):
    # Get YOLOv8 raw detections
    yx_boxes, conf, cls = run_yolov8(image_path, score_threshold=0.0, iou_threshold=0.7) # Run YOLOv8 to get raw detections
    if yx_boxes.shape[0] == 0: # If no boxes are detected
        return np.zeros((0,), np.float32), np.zeros((0,4), np.float32), np.zeros((0,), np.int32) # Return empty arrays

    N, C = yx_boxes.shape[0], len(class_names) # Get the number of boxes and classes
    box_conf = tf.convert_to_tensor(conf.reshape(1, N, 1), dtype=tf.float32) # Convert confidence scores to a TensorFlow tensor
    boxes    = tf.convert_to_tensor(yx_boxes.reshape(1, N, 4), dtype=tf.float32) # Convert bounding boxes to a TensorFlow tensor

    # One-hot class probabilities
    probs = np.zeros((1, N, C), dtype=np.float32) # Initialize a numpy array for class probabilities
    for i, c in enumerate(cls): # Iterate through class IDs
        if 0 <= c < C: # Check if the class ID is valid
            probs[0, i, c] = 1.0 # Set the probability for the predicted class to 1.0
    box_probs = tf.convert_to_tensor(probs, dtype=tf.float32) # Convert class probabilities to a TensorFlow tensor

    with Image.open(image_path) as im: # Open the image using PIL
        H, W = im.height, im.width # Get the image height and width

    scores, fboxes, classes = yolo_eval(
        box_conf, boxes, box_probs,
        image_shape=(H, W),
        max_boxes=max_boxes,
        score_threshold=score_threshold,
        iou_threshold=iou_threshold,
        per_class_nms=True
    ) # Apply YOLO evaluation with TensorFlow utilities

    # Optionally filter wanted class
    if wanted_class in class_names: # If a wanted class is specified
        idx = class_names.index(wanted_class) # Get the index of the wanted class
        mask = tf.where(classes == idx)[:,0] # Create a mask for the wanted class
        scores = tf.gather(scores, mask) # Gather scores for the wanted class
        fboxes = tf.gather(fboxes, mask) # Gather bounding boxes for the wanted class
        classes = tf.gather(classes, mask) # Gather class IDs for the wanted class

    return scores.numpy(), fboxes.numpy(), classes.numpy() # Return scores, bounding boxes, and class IDs as numpy arrays
````

### **What Are Anchors in YOLO?**

**Anchors** are predefined bounding box shapes (width × height) that act as *priors* to help YOLO detect objects of different sizes and aspect ratios.  
Instead of predicting bounding boxes from scratch, YOLO learns **offsets** relative to these anchors.

- Each grid cell on the feature map predicts multiple bounding boxes, one per anchor.  
- Anchors are chosen to reflect common object shapes in the training dataset (e.g., tall boxes for people, wide boxes for cars).  
- During training, YOLO matches ground truth boxes to the **closest anchor shape**, making learning more stable.  

**Why are anchors useful?**
- They let the model efficiently handle **multi-scale objects** in the same image.  
- They reduce the search space for bounding box regression, speeding up detection.  
- They improve accuracy by giving the network better "starting guesses" for box shapes.

In practice, YOLO uses 9 anchor boxes (in YOLOv3/v4) or different sets depending on the variant, tuned for the COCO dataset.  
Your `yolo_anchors.txt` file defines these anchor dimensions, which are loaded above and used during inference.

### Confidence & IoU Thresholds — What They Mean and What to Try

When you call `run_yolov8(image_path, score_threshold=0.25, iou_threshold=0.45)`, two key thresholds control what boxes you see:

1) Confidence (score) threshold  
- Meaning: Minimum predicted confidence for a box to be kept (objectness × class score).  
- Lower value (e.g., 0.25): keeps more boxes → more true positives but also more false positives.  
- Higher value (e.g., 0.60 or 0.90): keeps fewer boxes → fewer false positives but risks missing true objects.

2) IoU threshold (used in Non-Maximum Suppression, NMS)  
- Meaning: Intersection-over-Union overlap at which two boxes are considered “the same” object and the lower-scored one is suppressed.  
- Lower IoU (e.g., 0.45): more aggressive suppression → fewer overlapping boxes, may drop some valid close boxes.  
- Higher IoU (e.g., 0.75): less aggressive suppression → can keep more overlapping boxes, may show duplicates on the same object.

### Batch Processing a Folder of Images

So far, we have tested YOLO on a **single image**.  
The helper function `process_folder` extends this by running inference on **all images in a folder** and automatically producing a simple report.

What the function does:
1. Loops through each file in the specified `folder` (default = `"images"`).  
2. Runs `run_yolov8` to detect objects.  
3. Calls `draw_boxes` to visualize **only cars** (`wanted_idx=car`).  
4. Saves annotated images into an output folder (`out/`) if `save_out=True`.  
5. Builds a report listing the image filename and the number of cars detected (or an error message if inference failed).  

**Return Value:**  
A Python list of `(filename, car_count)` pairs, e.g.:
```python
[("0114.jpg", 3), ("0115.jpg", 5), ("0116.jpg", 0)]


````python
def process_folder(folder="images", min_score=0.25, save_out=True):
    out_dir = "out" # Define the output directory
    os.makedirs(out_dir, exist_ok=True) # Create the output directory if it doesn't exist
    report = [] # Initialize an empty list to store the report
    for img_file in glob.glob(os.path.join(folder, "*.*")): # Iterate through all files in the specified folder
        try:
            boxes, scores, classes = run_yolov8(img_file, score_threshold=min_score, iou_threshold=0.45) # Run YOLOv8 on the image
            vis, car_count = draw_boxes(img_file, boxes, scores, classes, class_names=class_names, wanted_idx=class_names.index("car")) # Draw boxes on the image, filtering for cars
            if save_out: # If save_out is True
                vis.save(os.path.join(out_dir, os.path.basename(img_file))) # Save the annotated image to the output directory
            report.append((os.path.basename(img_file), car_count)) # Append the filename and car count to the report
        except Exception as e: # Catch any exceptions
            report.append((os.path.basename(img_file), f"error: {e}")) # Append the filename and error message to the report
    return report # Return the report

# Example:
# counts = process_folder("images")
# counts
````

### Advanced Task 1: Class-wise Filtering and Comparison

````python
from collections import Counter # Import Counter from collections
import glob # Import glob
import matplotlib.pyplot as plt # Import matplotlib.pyplot as plt


# Multi-class comparison
target_classes = ["car", "person", "bus", "truck", "motorcycle", "zebra"] # Define the target classes
name_to_id = {v:k for k,v in model.names.items()} # Create a dictionary mapping class names to IDs

counts = Counter() # Initialize a Counter object

for img_file in glob.glob("/content/content/Lab 4/images/*.*"): # Iterate through all files in the images folder
    results = model.predict(img_file, conf=0.25, iou=0.45, verbose=False) # Run YOLOv8 prediction on the image
    r = results[0] # Get the first result
    if r.boxes is None: # If no boxes are detected
        continue # Skip to the next image
    cls_ids = r.boxes.cls.cpu().numpy().astype(int) # Get the class IDs of the detected objects
    for t in target_classes: # Iterate through the target classes
        if name_to_id[t] in cls_ids: # Check if the target class is present in the detected class IDs
            counts[t] += (cls_ids == name_to_id[t]).sum() # Count the number of detections for the target class

# Bar chart
plt.bar(counts.keys(), counts.values()) # Create a bar chart
plt.ylabel("Detections") # Set the y-axis label
plt.title("Class-wise detection counts") # Set the title of the chart
plt.show() # Display the chart

# Pie chart version
plt.pie(counts.values(), labels=counts.keys(), autopct="%1.1f%%") # Create a pie chart
plt.title("Class-wise detection distribution") # Set the title of the chart
plt.show() # Display the chart
````

### Advanced Task 2: Threshold Sweep with Detection Counts

You already tried changing confidence manually.  
Now automate it:
- Sweep across confidence thresholds (0.1 → 0.9).  
- Count how many cars are detected at each threshold.  
- Plot detections vs threshold as a line chart.

**Deliverable:** A plot showing the trade-off, and a sentence describing the “sweet spot” threshold for your dataset.


````python
conf_list = [0.1,0.2,0.3,0.5,0.7,0.9] # Define a list of confidence thresholds
counts = [] # Initialize an empty list to store counts

for conf in conf_list: # Iterate through the confidence thresholds
    total = 0 # Initialize a total count for the current threshold
    for img_file in glob.glob("/content/content/Lab 4/images/*.*"): # Iterate through all files in the images folder
        boxes, scores, classes = run_yolov8(img_file, score_threshold=conf, iou_threshold=0.45) # Run YOLOv8 with the current confidence threshold
        total += (classes == class_names.index("car")).sum() # Count the number of cars detected
    counts.append(total) # Append the total count to the counts list

plt.plot(conf_list, counts, marker="o") # Create a line plot
plt.xlabel("Confidence Threshold") # Set the x-axis label
plt.ylabel("Number of Cars Detected") # Set the y-axis label
plt.title("Car detections vs Confidence Threshold") # Set the title of the chart
plt.grid(True) # Display a grid
plt.show() # Display the chart
````
